# Fill NaN dengan mean yang di-floor untuk non object dan isi dengan modus untuk object

```python
import math

def fill_value(df):
    '''
    Isi setiap value yang null
    Jika kolomnya bertipe Numerik maka isi dengan mean
    lain kolomnya bertipe Kategorikal maka isi dengan nilai yang frekuensinya paling banyak
    '''

    '''
    tahun_kelahiran              int64
    pendidikan                  object
    status_pernikahan           object
    pendapatan                 float64
    jumlah_anak_balita         float64
    jumlah_anak_remaja         float64
    terakhir_belanja           float64
    belanja_buah               float64
    belanja_daging             float64
    belanja_ikan               float64
    belanja_kue                float64
    pembelian_diskon           float64
    pembelian_web              float64
    pembelian_toko             float64
    keluhan                    float64
    tanggal_menjadi_anggota     object
    '''

    df['tahun_kelahiran'] = df['tahun_kelahiran'].fillna(math.floor(df['tahun_kelahiran'].mean()))
    df['pendidikan'] = df['pendidikan'].replace('5', pd.NA) # Mengganti nilai '5' dengan NaN
    df['pendidikan'] = df['pendidikan'].fillna(df['pendidikan'].mode()[0]) # Mengisi nilai NaN dengan modus
    df['status_pernikahan'] = df['status_pernikahan'].replace('5', pd.NA)
    df['status_pernikahan'] = df['status_pernikahan'].fillna(df['status_pernikahan'].mode()[0])
    df['pendapatan'] = df['pendapatan'].fillna(math.floor(df['pendapatan'].mean()))
    df['jumlah_anak_balita'] = df['jumlah_anak_balita'].fillna(math.floor(df['jumlah_anak_balita'].mean()))
    df['jumlah_anak_remaja'] = df['jumlah_anak_remaja'].fillna(math.floor(df['jumlah_anak_remaja'].mean()))
    df['terakhir_belanja'] = df['terakhir_belanja'].fillna(math.floor(df['terakhir_belanja'].mean()))
    df['belanja_buah'] = df['belanja_buah'].fillna(math.floor(df['belanja_buah'].mean()))
    df['belanja_daging'] = df['belanja_daging'].fillna(math.floor(df['belanja_daging'].mean()))
    df['belanja_ikan'] = df['belanja_ikan'].fillna(math.floor(df['belanja_ikan'].mean()))
    df['belanja_kue'] = df['belanja_kue'].fillna(math.floor(df['belanja_kue'].mean()))
    df['pembelian_diskon'] = df['pembelian_diskon'].fillna(math.floor(df['pembelian_diskon'].mean()))
    df['pembelian_web'] = df['pembelian_web'].fillna(math.floor(df['pembelian_web'].mean()))
    df['pembelian_toko'] = df['pembelian_toko'].fillna(math.floor(df['pembelian_toko'].mean()))
    df['keluhan'] = df['keluhan'].fillna(math.floor(df['keluhan'].mean()))
    df['tanggal_menjadi_anggota'] = df['tanggal_menjadi_anggota'].fillna(df['tanggal_menjadi_anggota'].mode()[0])
```

# Normalisasi

```python
def normalize(df):
    # Normalisasi setiap kolom yang bertipe numerik
    # Gunakan MinMaxScaler
    from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder, StandardScaler
    scaler = MinMaxScaler()
    ordinal = OrdinalEncoder()
    df[['tahun_kelahiran', 'pendapatan', 'jumlah_anak_balita', 'jumlah_anak_remaja', 'terakhir_belanja', 'belanja_buah', 'belanja_daging', 'belanja_ikan', 'belanja_kue', 'pembelian_diskon', 'pembelian_web', 'pembelian_toko', 'keluhan']] = scaler.fit_transform(df[['tahun_kelahiran', 'pendapatan', 'jumlah_anak_balita', 'jumlah_anak_remaja', 'terakhir_belanja', 'belanja_buah', 'belanja_daging', 'belanja_ikan', 'belanja_kue', 'pembelian_diskon', 'pembelian_web', 'pembelian_toko', 'keluhan']])
    df[['pendidikan', 'status_pernikahan']]=ordinal.fit_transform(df[['pendidikan', 'status_pernikahan']])
```

# Split Data

```python
from sklearn.model_selection import train_test_split

y = pd.read_csv('data-analytics-competition-find-it-2024/train_labels.csv')
y
```

```python
y['jumlah_promosi'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.05)
```

# Modelling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

dt = DecisionTreeClassifier(random_state=24)
dt.fit(X_train, y_train)

prediction = dt.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=24)
rf.fit(X_train, y_train)

prediction = rf.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear', random_state=24)
svc.fit(X_train, y_train)

prediction = svc.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(random_state=24)
gbc.fit(X_train, y_train)

prediction = gbc.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

prediction = knn.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(random_state=24)
mlp.fit(X_train, y_train)

prediction = mlp.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=24)
logreg.fit(X_train, y_train)

prediction = logreg.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)

prediction = nb.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(random_state=24)
xgb_model.fit(X_train, y_train)

prediction = xgb_model.predict(X_test)
print(f1_score(prediction, y_test, average='macro'))
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(7, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), verbose=2)

prediction = model.predict(X_test)
prediction_classes = prediction.argmax(axis=-1)
print(f1_score(prediction_classes, y_test, average='macro'))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_rfr = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'max_features': ['sqrt'],
    'random_state': [42]
}

param_grid_xg = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [50, 100, 150],
}

grid_rfr = GridSearchCV(
    RandomForestClassifier(),
    param_grid=param_grid_rfr,
    n_jobs=-1,
    cv=2,
    verbose=1
)

grid_xg = GridSearchCV(
    estimator=xgb_model, 
    param_grid=param_grid_xg,  
    cv=5)
```

```python
grid_rfr.fit(X_train, y_train)

best_rfr_model = grid_rfr.best_estimator_
best_rfr_params = grid_rfr.best_params_

y_pred_rfr = best_rfr_model.predict(X_test)

print(f1_score(y_test, y_pred_rfr, average='macro'))
```

```python
grid_xg.fit(X_train, y_train)

best_xg_model = grid_xg.best_estimator_
best_xg_params = grid_xg.best_params_

y_pred_xg = best_xg_model.predict(X_test)

print(f1_score(y_test, y_pred_xg, average='macro'))
```
