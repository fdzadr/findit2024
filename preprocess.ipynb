```python
# Import Library
import pandas as pd
print(pd.__version__)
```

```python
# Import Data
df = pd.read_csv('data-analytics-competition-find-it-2024/train_features.csv')
df
```

```python
# Tampilkan jenis setiap kolom
df.dtypes
```

```python
def fill_value():
    # Isi setiap value yang null
    # Jika kolomnya bertipe Numerik maka isi dengan mean
    # lain kolomnya bertipe Kategorikal maka isi dengan nilai yang frekuensinya paling banyak
    '''
    Mean, Median, and Mode:

    Mean - The average value --> Pakai ini untuk numerik
    Median - The mid point value
    Mode - The most common value --> Pakai ini untuk kategorikal
    '''

    '''
    tahun_kelahiran              int64
    pendidikan                  object
    status_pernikahan           object
    pendapatan                 float64
    jumlah_anak_balita         float64
    jumlah_anak_remaja         float64
    terakhir_belanja           float64
    belanja_buah               float64
    belanja_daging             float64
    belanja_ikan               float64
    belanja_kue                float64
    pembelian_diskon           float64
    pembelian_web              float64
    pembelian_toko             float64
    keluhan                    float64
    tanggal_menjadi_anggota     object
    '''

    df['tahun_kelahiran'] = df['tahun_kelahiran'].fillna(df['tahun_kelahiran'].mean())
    df['pendidikan'] = df['pendidikan'].fillna(df['pendidikan'].mode()[0]) #  [0] digunakan untuk mengekstrak elemen pertama dari Seri
    df['status_pernikahan'] = df['status_pernikahan'].fillna(df['status_pernikahan'].mode()[0])
    df['pendapatan'] = df['pendapatan'].fillna(df['pendapatan'].mean())
    df['jumlah_anak_balita'] = df['jumlah_anak_balita'].fillna(df['jumlah_anak_balita'].mean())
    df['jumlah_anak_remaja'] = df['jumlah_anak_remaja'].fillna(df['jumlah_anak_remaja'].mean())
    df['terakhir_belanja'] = df['terakhir_belanja'].fillna(df['terakhir_belanja'].mean())
    df['belanja_buah'] = df['belanja_buah'].fillna(df['belanja_buah'].mean())
    df['belanja_daging'] = df['belanja_daging'].fillna(df['belanja_daging'].mean())
    df['belanja_ikan'] = df['belanja_ikan'].fillna(df['belanja_ikan'].mean())
    df['belanja_kue'] = df['belanja_kue'].fillna(df['belanja_kue'].mean())
    df['pembelian_diskon'] = df['pembelian_diskon'].fillna(df['pembelian_diskon'].mean())
    df['pembelian_web'] = df['pembelian_web'].fillna(df['pembelian_web'].mean())
    df['pembelian_toko'] = df['pembelian_toko'].fillna(df['pembelian_toko'].mean())
    df['keluhan'] = df['keluhan'].fillna(df['keluhan'].mean())
    df['tanggal_menjadi_anggota'] = df['tanggal_menjadi_anggota'].fillna(df['tanggal_menjadi_anggota'].mode()[0])
```

```python
def normalize():
    # Normalisasi setiap kolom yang bertipe numerik
    # Gunakan MinMaxScaler
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    df[['tahun_kelahiran', 'pendapatan', 'jumlah_anak_balita', 'jumlah_anak_remaja', 'terakhir_belanja', 'belanja_buah', 'belanja_daging', 'belanja_ikan', 'belanja_kue', 'pembelian_diskon', 'pembelian_web', 'pembelian_toko', 'keluhan']] = scaler.fit_transform(df[['tahun_kelahiran', 'pendapatan', 'jumlah_anak_balita', 'jumlah_anak_remaja', 'terakhir_belanja', 'belanja_buah', 'belanja_daging', 'belanja_ikan', 'belanja_kue', 'pembelian_diskon', 'pembelian_web', 'pembelian_toko', 'keluhan']])
```

```python
fill_value()
df
```

```python
normalize()
df
```
